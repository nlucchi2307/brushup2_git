```python
import gspread
from gspread_dataframe import get_as_dataframe
from google.colab import auth
from oauth2client.client import GoogleCredentials
from google.auth import default
import pandas as pd
```

```python
# Authenticate and create a client
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# we take an excel spreadsheet with mutiple worksheet tabs, upload it to the drive and open it in google sheets.
# then paste the link to that sheet here:
spreadsheet_url = 'https://docs.google.com/spreadsheets/d/1RWcuf025aIeie6rfXYCnXvNdZW9iHjTUhupdQuA7xLI/edit?gid=1578783303#gid=1578783303'
```

```python
# Open the spreadsheet
spreadsheet = gc.open_by_url(spreadsheet_url)
```

```python
# Get all the worksheet objects (tabs) in the spreadsheet
worksheets = spreadsheet.worksheets()
```

```python
# Create a dictionary of DataFrames for each sheet - NB: we're using a spreadsheet that
# only has one sheet but the functioning is the same.

dfs = {}
for sheet in worksheets:
    sheet_name = sheet.title
    df = get_as_dataframe(sheet)
    dfs[sheet_name] = df
```

```python
# Print the names of the sheet (same code if we had more sheets)
for sheet_name, df in dfs.items():
    print(f"name of sheet: {sheet_name}")
```

```python
# Show first few rows of the DataFrame (same code if we had more sheets)
print(df.head())
```

```python
# this code must be run for each sheet: df2, df3... and between square brackets in the
# first line we put the 'name of sheet' that results from the code in cell 18
df1 = dfs['WB_pop_clean']
print('initial shape:', df1.shape)
# we need to convert column names in strings in order to use str.match
df1.columns = df1.columns.astype(str)
df1 = df1.loc[:, ~df1.columns.str.match("Unnamed")]
print('after removing unnamed columns: ', df1.shape)
df1 = df1.dropna()
print('after removing rows with NaNs: ', df1.shape)
df1
```
